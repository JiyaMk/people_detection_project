capturing images using a video


```python
cap = cv2.VideoCapture("/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/videoplayback.mp4")



import os
directory = "/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/All_Frames"
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
frame_number = 0
while True:
    ret, frame = cap.read()
    if not ret:
        break
    image_path = os.path.join(directory, f"frame_{frame_number:04d}.jpg")
    cv2.imwrite(image_path, frame)

    frame_number += 1
cap.release()
```

```python
import cvlib as cv
from cvlib.object_detection import draw_bbox
import matplotlib.pyplot as plt
import csv
```

Counting the no. of person in a frame and storing the count in a csv file in front of each frame

```python
def detect_people_count(image_path):
    img = cv2.imread(image_path)
    box, label, c_score = cv.detect_common_objects(img)
    num_people = label.count('person')
    return num_people
input_folder = '/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/All_Frames'

image_files = [f for f in os.listdir(input_folder) if f.endswith('.jpg')]

output_file = '/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/labels.csv'

with open(output_file, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)

    csv_writer.writerow(['Image', 'No. of People'])

    for image_file in image_files:
        image_path = os.path.join(input_folder, image_file)
        num_people = detect_people_count(image_path)
        csv_writer.writerow([image_file, num_people])
```

converting the images to NumPy array

```python
inp_folder = "/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/All_Frames"
image_files = [f for f in os.listdir(inp_folder) if f.endswith('.jpg')]

images_list = []

for image_file in image_files:
    image_path = os.path.join(inp_folder, image_file)
    image = cv2.imread(image_path)
    img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    images_list.append(img)
output_npy_file = "/content/drive/MyDrive/Colab Notebooks/DATASET_PERSON_DETECTION/All_Frames.npy"
np.save(output_npy_file, images_list)
```
